### Convertendo os tipos de dados 

```python


date_cols = ["recorddate_key", "birthdate_key", "orighiredate_key", "terminationdate_key"] 
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")


str_cols = ["city_name", "department_name", "job_title",
            "gender_short", "gender_full",
            "termreason_desc", "termtype_desc",
            "STATUS", "BUSINESS_UNIT"]

for col in str_cols:
    df[col] = df[col].astype("string")


print(df.dtypes)
```

### Excluindo colunas redundantes 

```python
cols_drop = ['age', 'length_of_service', 'STATUS_YEAR', 'gender_full', 'STATUS']

df = df.drop(columns=cols_drop)

```

```python
df.columns.tolist()
```

### Tratando a data de recisão

```python
df.loc[df['terminationdate_key'] == "1900-01-01", 'terminationdate_key'] = pd.NaT
```

### Padronizando strings

```python
str_cols = df.select_dtypes(include="string").columns

for col in str_cols:
    df[col] = df[col].str.strip().str.title()
```
